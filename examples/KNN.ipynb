# **KNN. Classification/Regression**

```python
from sklearn.datasets import load_breast_cancer, fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from scratchml.supervised_learning.clustering import KNNClassifier
```

## **Classification**

### **Data**

We will be using the https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html dataset to test the model.

```python
# Load the California Housing dataset
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
```

```python
# Standardizing the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA
pca = PCA(n_components=2)  # n_components is the number of principal components to keep
X_pca = pca.fit_transform(X_scaled)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Print the shapes of the resulting sets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

### **Models fitting**

```python
# Insantiate models
clf_KNN = KNNClassifier()

# Fit models
clf_KNN.visualize_classification(X_train, y_train, X_test, y_test)
```

### **Evaluation (Accuracy score)**

```python
# Evaluate models
y_pred = clf_KNN.predict(X_test, X_train, y_train)
acc_KNN = accuracy_score(y_test, y_pred)

print(f"Acc for KNN (scratchml): {acc_KNN}")
```
